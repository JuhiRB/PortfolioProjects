```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else: 
        return sex
```

```python
titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis =1)
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []
for level in deck:
      levels.append(level[0])  
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.catplot(x='Cabin',kind = 'count',data = cabin_df ,palette = 'winter')
```

```python
cabin_df =cabin_df[cabin_df.Cabin != 'T']

sns.catplot(x='Cabin',kind = 'count',data = cabin_df ,palette = 'summer')
```

```python
#Who was alone and who was with family
#Sib Sp = If any sibling in on board
#parch = If Parent or child is on board
titanic_df.head()
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch

titanic_df['Alone']
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'

titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no',1:'yes'})

sns.catplot(x='Survivor',data=titanic_df,palette='Set1',kind='count')
```
